```python
## we have to change the unit price data type to float or int but there is a $ sign ,unable to identify it so first replace this empty space
# $ sign ko hata kar changes ko wapas usi column mein save karne ke liye:
df['unit_price'] = df['unit_price'].str.replace('$', '', regex=False)
```

```python
# convert into float data type
df['unit_price'] = df['unit_price'].astype(float)

# New column 
df['total'] = df['unit_price'] * df['quantity']
```
